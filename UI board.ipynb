```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
com_list =[]

def spider(max_pages):
    page = 1
    while page < max_pages:
        url = "http://www.ui-line.com/html/cust/cust01/cust_01_02.php?&nowPage=" + str(page)
        html = urlopen(url)
        soup = BeautifulSoup(html,"lxml")
        titles = soup.find_all("td", {"class":"tl"})
        for title in titles:
            com_list.append(title.get_text())
        page+=1
        if page%100 == 0:
            print(page)
```

```python
spider(1010)
```

```python
for i in com_list:
    print(i)
```
